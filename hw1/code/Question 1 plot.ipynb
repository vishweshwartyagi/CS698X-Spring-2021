```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import laplace
```

```python
eta = 1
gamma = 1
plt.figure(figsize = (12,10))
x = np.linspace(-4, 4, 100)
plt.plot()
plt.plot(x, stats.norm.pdf(x, 0, 1/eta))
plt.plot(x, laplace.pdf(x, 0, 1/gamma), label = '$p(x\;|\;\gamma)$')
plt.title('$\eta = 1 \;\; \gamma = 1$')
plt.savefig('images/Gaussian and Laplace.png')
plt.legend()
plt.show()
```
